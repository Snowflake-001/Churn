Finding the mean of Churn Customers in each state using groupby function

```python
Telecom.groupby('State')['Churn'].mean().round(4)*100
```

Line chart showing State-wise Churn rate

```python
x=Telecom['State'].unique()
y=Telecom.groupby('State')['Churn'].mean()
sns.set(style='darkgrid')
plt.rcParams['figure.figsize']=(20,7)
plt.rcParams['lines.linestyle']=('-')
plt.plot(x,y, color='black', marker='*', markersize='10')
plt.title('State Churn Graph', color='maroon', fontsize=25)
plt.xlabel('States',color='purple', fontsize=20)
plt.ylabel('Churn Rate', color='purple', fontsize=20)
```

Crosstab showing State vs Churn percentage

```python
c=pd.crosstab(Telecom['State'], Telecom['Churn'], margins=True)
c['Percentage_Churn']=c[True]/(c['All'])*100
print(c.head())
print(type(c))
```

```python
c.info()
```

```python
c.sort_values(by='Percentage_Churn', ascending=False).head()
```

Analyzing Per Minute Charge

```python
Telecom.head(2)
```

```python
day_pm=Telecom['Total day charge'].mean()/Telecom['Total day minutes'].mean()
eve_pm=Telecom['Total eve charge'].mean()/Telecom['Total eve minutes'].mean()
nigth_pm=Telecom['Total night charge'].mean()/Telecom['Total night minutes'].mean()
int_pm=Telecom['Total intl charge'].mean()/Telecom['Total intl minutes'].mean()

print(day_pm)
print(eve_pm)
print(nigth_pm)
```

Bar plot

```python
categories=['day', 'evening', 'night', 'int']
sns.barplot(x=categories, y=[day_pm, eve_pm, nigth_pm, int_pm], hue=categories, palette=['blue', 'orange', 'green', 'brown'])
plt.show()
```
